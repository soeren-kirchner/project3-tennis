# Train the Multi-Agent
## Normal Noise 

```python
multi_agent = MultiAgent(num_agents, state_size, action_size, noise_type="normal")
scores_normal = train(multi_agent, n_episodes=1300, max_t=10000)
```

```python
plot_scores([scores_normal], ['normal'], save_as='normal.png')
```

## Ornstein–Uhlenbeck Process Action Noise

```python
multi_agent = MultiAgent(num_agents, state_size, action_size, noise_type="ou")
scores_ou = train(multi_agent, n_episodes=1300, max_t=10000)
```

```python
plot_scores([scores_ou], ['ou'], save_as='ou.png')
```

## Parameter Action Noise

```python
multi_agent = MultiAgent(num_agents, state_size, action_size, noise_type="param")
scores_param = train(multi_agent, n_episodes=1300, max_t=10000)
```

```python
plot_scores([scores_param], ['param'], save_as='param.png')
```

## Comparison

```python
scores_normal_means = np.asarray(scores_normal).T[1]
scores_ou_means = np.asarray(scores_ou).T[1]
scores_param_means = np.asarray(scores_param).T[1]
plot_scores([scores_ou_means, scores_normal_means, scores_param_means], ['ou', 'normal', 'param'], save_as='compare.png')
```
